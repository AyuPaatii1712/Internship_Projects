```python
rating_count = rating['bookrating'].value_counts()
rating_count
```

```python
x = rating['ISBN'].value_counts()>50
x[x]
```

```python
rat = rating.drop('userid', axis=1)
rat
```

```python
def lst(x):
    if all(itm == 0 for itm in x):
        return x
    else:
        lt = [i for i in x if i != 0]
        return lt
```

```python
rat = rating.drop('userid', axis=1)
rate = pd.DataFrame(rat.groupby('ISBN')['bookrating'].apply(list)).reset_index()
rate['bookrating1'] = rate['bookrating'].apply(lambda x: lst(x))
rate
```

```python
rate['counts'] = rate['bookrating'].apply(len)
rate['counts_no_zero'] = rate['bookrating1'].apply(len)
rate
```

```python
rate['avg_rating'] = rate['bookrating'].apply(st.mean)
rate['avg_rating_no_zero'] = rate['bookrating1'].apply(st.mean)
rate['final_rating'] = round((rate['avg_rating']+rate['avg_rating_no_zero'])/2 , 3)
rate = rate.sort_values('counts', ascending=False).reset_index()
rate
```

```python
def country_popular_books(user, rating, book, location):
    user_country=user[user['location']==location]
    df=user_usa.merge(rating,on='userid')
    df_new=df[['ISBN','bookrating']]
    y = rating['ISBN'].value_counts()>50
    country_rate = pd.DataFrame(df_new.groupby('ISBN')['bookrating'].apply(list)).reset_index()
    country_rate['bookrating1'] = country_rate['bookrating'].apply(lambda x: lst(x))
    country_rate['counts'] = country_rate['bookrating'].apply(len)
    country_rate['counts_no_zero'] = country_rate['bookrating1'].apply(len)
    country_rate['avg_rating'] = country_rate['bookrating'].apply(st.mean)
    country_rate['avg_rating_no_zero'] = country_rate['bookrating1'].apply(st.mean)
    country_rate['final_rating'] = round((country_rate['avg_rating']+country_rate['avg_rating_no_zero'])/2 , 3)
    country_rate = country_rate.sort_values('counts', ascending=False).reset_index()
    loc_rate_country=country_rate.merge(book,on='ISBN')
    loc_with_rate=loc_rate_country[['ISBN','title','counts','counts_no_zero','final_rating']]
    populer_country=loc_with_rate[loc_with_rate['counts_no_zero']>=150].sort_values('final_rating',ascending=False).reset_index()
    
    return populer_country
```

```python
popular_usa = country_popular_books(user, rating, book, ' usa')
popular_usa.head(20)
```
